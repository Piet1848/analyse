```python
import numpy as np
import matplotlib.pyplot as plt
import os

import data_organizer as do
import run_evaluation
from calculator import Calculator, cornell_potential_ansatz

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
import notebook_analysis_helpers as nah
```

```python
TARGET_RUN_PATH = "../data/20260316/42"
SEARCH_ROOT = "../data/20260316/"

THERMALIZATION_STEPS = 500
SOMMER_TARGET_FORCE = 1.65

print(f"Target Run: {TARGET_RUN_PATH}")
print(f"Searching for matching runs in: {SEARCH_ROOT}")
print(f"Parameters: Therm={THERMALIZATION_STEPS}, Target={SOMMER_TARGET_FORCE}")
```

# Potential V(R)

```python
# Find all runs with the same parameters and combine their W_temp.out files

from pathlib import Path

try:
    # 1. Get the group key for the target run
    target_key = run_evaluation._group_key_for_run(TARGET_RUN_PATH)
    if target_key is None:
        raise ValueError(f"Could not load parameters from {TARGET_RUN_PATH}. Is input.yaml present?")
    
    print(f"Target run group key: {target_key}")
    
    # 2. Find all matching runs in the search root
    matching_runs = []
    for dirpath, dirnames, filenames in os.walk(SEARCH_ROOT):
        if "input.yaml" in filenames:
            key = run_evaluation._group_key_for_run(dirpath)
            if key == target_key:
                matching_runs.append(os.path.abspath(dirpath))
                
    matching_runs = sorted(set(matching_runs))
    print(f"Found {len(matching_runs)} matching runs:")
    for m in matching_runs:
        print(f"  - {m}")
        
    # 3. Load and combine W_temp.out from all matching runs
    w_temp_files = []
    for run_dir in matching_runs:
        full_path = os.path.join(run_dir, "W_temp.out")
        if os.path.exists(full_path):
            fd = do.FileData(full_path)
            fd.read_file()
            w_temp_files.append(fd)
        else:
            print(f"Warning: W_temp.out not found in {full_path}")
            
    if not w_temp_files:
        print("Error: No W_temp.out files found in the specified directories.")
    else:
        print(f"Combining {len(w_temp_files)} W_temp.out files...")
        combined_w_temp = do.combine_file_data(w_temp_files, min_step=THERMALIZATION_STEPS, source_name="combined_W_temp")
       # Free up RAM by deleting the individual file objects
        del w_temp_files
        
        if combined_w_temp is None:
            print("Error: Failed to combine file data.")
        else:
            w_temp_data = combined_w_temp # Required for subsequent cells
            print(f"Successfully combined data. Total entries: {len(combined_w_temp.observables[0].values)}")
            calc = Calculator(combined_w_temp, n_bootstrap=50, step_size=1)
            print("Calculator initialized with combined data.")

except Exception as e:
    print(f"An error occurred loading and combining data: {e}")
```

```python

# Shared notebook-first cache/dashboard workflow
notebook_scan_cache = nah.load_or_build_scan_cache(
    matching_runs,
    thermalization_steps=THERMALIZATION_STEPS,
    target_force=SOMMER_TARGET_FORCE,
    n_bootstrap=run_evaluation.DEFAULT_N_BOOTSTRAP,
    block_size=run_evaluation.DEFAULT_BLOCK_SIZE,
    load_workers=LOAD_WORKERS if 'LOAD_WORKERS' in globals() else 1,
)
notebook_dashboard_state = nah.build_notebook_dashboard_state(notebook_scan_cache)
notebook_dashboard_state.get("dashboard", notebook_scan_cache.get("audit"))
```

```python
# Plot W(R, T) vs T

# Identify available R values (L column in W_temp)
try:
    all_L = np.asarray(w_temp_data.get("L").values)
    unique_Rs = sorted(np.unique(all_L))
    
    # Pick a few representative R values to plot
    # e.g., smallest, middle, largest
    r_to_plot = [unique_Rs[0], unique_Rs[len(unique_Rs)//2], unique_Rs[-1]]
    if len(unique_Rs) > 3:
        # Maybe pick explicitly R=2, 4, 6 if available
        candidates = [2, 4, 6, 8]
        r_to_plot = [r for r in candidates if r in unique_Rs]
        if not r_to_plot: r_to_plot = unique_Rs[:3]

    # Define a set of t_min values to compare the fit stability
    t_min_values = np.arange(1, 10)
    # Define a maxT value that can be tuned and changed (None means no upper limit)
    t_max_value = None
    
    print(f"Plotting W(R,T) fits for R = {r_to_plot}")
    print(f"Comparing fits for t_min = {t_min_values} and t_max = {t_max_value}")

    fig, axes = plt.subplots(1, len(r_to_plot), figsize=(5*len(r_to_plot), 5), constrained_layout=True)
    if len(r_to_plot) == 1: axes = [axes]

    for ax, r_val in zip(axes, r_to_plot):
        # 1. Get W(R, T) data points
        # We need to query T values for this R
        all_T = np.asarray(w_temp_data.get("T").values)
        unique_Ts = sorted(np.unique(all_T[all_L == r_val]))
        
        ws, ws_err, ts = [], [], []
        
        for t_val in unique_Ts:
            try:
                # W_R_T calculator retrieves the mean and error for W(R,T)
                w_var = calc.get_variable("W_R_T", R=r_val, T=t_val)
                ws.append(w_var.get())
                ws_err.append(w_var.err() if w_var.err() else 0)
                ts.append(t_val)
            except Exception:
                continue
                
        # Filter negative or zero values for log plot
        ts = np.array(ts, dtype=np.float32)
        ws = np.array(ws, dtype=np.float32)
        ws_err = np.array(ws_err, dtype=np.float32)
        
        mask = ws > 0
        ts_fit = ts[mask]
        ws_fit = ws[mask]
        ws_err_fit = ws_err[mask]
        
        # Plot Data Points
        ax.errorbar(ts_fit, ws_fit, yerr=ws_err_fit, fmt='o', color='black', alpha=0.3, label=f'Data R={r_val}')
        
        colors = plt.cm.viridis(np.linspace(0, 0.9, len(t_min_values)))
        
        for idx, t_start in enumerate(t_min_values):
            try:
                v_var = calc.get_variable("V_R", R=r_val, t_min=t_start, t_max=t_max_value)
                v_val = v_var.get()
                
                C_est = v_var.parameters.get("fit_C", None)
                
                if not np.isnan(v_val):
                    if C_est is None:
                        # Fallback if fit_C is missing
                        if t_max_value is not None:
                            mask_t = (ts_fit >= t_start) & (ts_fit <= t_max_value)
                        else:
                            mask_t = ts_fit >= t_start
                        if np.any(mask_t):
                            ln_ws = np.log(ws_fit[mask_t])
                            ts_subset = ts_fit[mask_t]
                            ln_C_vals = ln_ws + v_val * ts_subset
                            ln_C = np.mean(ln_C_vals)
                            C_est = np.exp(ln_C)
                        else:
                            C_est = 1.0

                    t_line = np.linspace(min(ts_fit), max(ts_fit), 100)
                    w_line = C_est * np.exp(-v_val * t_line)
                    
                    ax.plot(t_line, w_line, '-', color=colors[idx], label=f't_min={t_start}: V={v_val:.4f}')
                    
                    # Highlight fit region start
                    # ax.axvline(t_start, color=colors[idx], linestyle=':', alpha=0.5)

            except Exception as e:
                print(f"Could not get V_R for R={r_val}, t_min={t_start}, t_max={t_max_value}: {e}")

        ax.set_yscale('log')
        ax.set_title(f"Wilson Loop Decay (R={r_val})")
        ax.set_xlabel("T")
        ax.set_ylabel("W(R,T)")
        ax.legend()

    plt.show()

except Exception as e:
    print(f"Error in plotting W(R,T): {e}")
```

```python
# Plot the effective mass
try:
    print("Calculating Effective Mass...")
    
    # Use the same R values as before
    r_to_plot_eff = r_to_plot if 'r_to_plot' in locals() else unique_Rs[:3]
    
    # Determine max T available
    # Assuming 'all_T' is available from previous cells
    max_T = int(np.max(all_T))
    # We can only go up to max_T - 1 because we need T+1
    t_range = np.arange(1, max_T) 
    
    fig, axes = plt.subplots(1, len(r_to_plot_eff), figsize=(5*len(r_to_plot_eff), 5), constrained_layout=True)
    if len(r_to_plot_eff) == 1: axes = [axes]
    
    for ax, r_val in zip(axes, r_to_plot_eff):
        eff_masses = []
        eff_errs = []
        ts_eff = []
        
        for t_val in t_range:
            
            try:
                # Calculate effective mass at T. 
                # Note: effective_mass(R, T) computes ln(W(R,T)/W(R,T+1))
                m_var = calc.get_variable("effective_mass", R=r_val, T=t_val)
                val = m_var.get()
                err = m_var.err()
                
                if not np.isnan(val) and np.isfinite(val): # Check for validity
                    eff_masses.append(val)
                    eff_errs.append(err if err is not None else 0)
                    ts_eff.append(t_val)
            except Exception:
                continue
                
        if not ts_eff:
            print(f"No effective mass data for R={r_val}")
            continue

        # Plot effective mass points
        # Ideally, plateaus should match the V(R) extracted from fits
        ax.errorbar(np.array(ts_eff) + 0.5, eff_masses, yerr=eff_errs, fmt='o', label=f'Eff. Mass R={r_val}')
        
        # Overlay the potential V(R) obtained from the fit (using t_min=R0_T_MIN)
        # Assuming R0_T_MIN is defined in previous cells (it was 5 in the provided context)
        t_min_fit = R0_T_MIN if 'R0_T_MIN' in locals() else 5
        t_max_fit = t_max_value if 't_max_value' in locals() else None
        try:
            v_var = calc.get_variable("V_R", R=r_val, t_min=t_min_fit, t_max=t_max_fit)
            v_val = v_var.get()
            v_err = v_var.err()
            
            if not np.isnan(v_val):
                 ax.axhline(v_val, color='r', linestyle='--', label=f'Fit V(R) (t_min={t_min_fit})')
                 if v_err is not None:
                     ax.fill_between([min(ts_eff), max(ts_eff)+1], v_val - v_err, v_val + v_err, color='r', alpha=0.1)
        except Exception:
            pass

        ax.set_title(f"Effective Mass (R={r_val})")
        ax.set_xlabel("T + 0.5")
        ax.set_ylabel("$m_{eff}(T)$")
        ax.legend()
        
    plt.show()

except Exception as e:
    print(f"Error plotting effective mass: {e}")
```

```python
try:
    fig, ax = plt.subplots(figsize=(10, 6))

    # Compare a few t_min values to check for plateau stability
    comparison_t_mins = t_min_values

    for t_start in comparison_t_mins:
        rs_plot, vs_plot, err_plot = [], [], []
        
        for r_val in unique_Rs:
            try:
                # Extract V(R) for this specific t_min
                v_var = calc.get_variable("V_R", R=r_val, t_min=t_start, t_max=t_max_value if 't_max_value' in locals() else None)
                val = v_var.get()
                
                if not np.isnan(val):
                    rs_plot.append(r_val)
                    vs_plot.append(val)
                    err_plot.append(v_var.err() if v_var.err() is not None else 0)
            except Exception:
                continue
        
        ax.errorbar(rs_plot, vs_plot, yerr=err_plot, fmt='o-', capsize=3, label=f'$t_{{min}}={t_start}$')

    ax.set_title("Static Potential $V(R)$ Stability (Comparison of $t_{min}$)")
    ax.set_xlabel("$R/a$")
    ax.set_ylabel("$aV(R)$")
    ax.legend()
    plt.show()

except Exception as e:
    print(f"Error comparing V(R) fits: {e}")

    
try:
    target_R = 2
    if target_R in unique_Rs:
        t_mins_r6, vs_r6, errs_r6 = [], [], []
        
        for t_start in t_min_values:
            try:
                # Extract V(R) for the target R and current t_min
                v_var = calc.get_variable("V_R", R=target_R, t_min=t_start, t_max=t_max_value if 't_max_value' in locals() else None)
                val = v_var.get()
                    
                if not np.isnan(val):
                    t_mins_r6.append(t_start)
                    vs_r6.append(val)
                    errs_r6.append(v_var.err() if v_var.err() is not None else 0)
            except Exception:
                continue
            
        fig, ax = plt.subplots(figsize=(8, 5))
        ax.errorbar(t_mins_r6, vs_r6, yerr=errs_r6, fmt='s-', capsize=4, color='firebrick', label=f'$R={target_R}$')
           
        ax.set_title(f"Plateau Stability: $V(R={target_R})$ vs. $t_{{min}}$")
        ax.set_xlabel("$t_{min}$")
        ax.set_ylabel(f"$aV(R={target_R})$")
        ax.legend()
        plt.show()
    else:
        print(f"R={target_R} not found in unique_Rs.")

except Exception as e:
    print(f"Error plotting V(R=6) stability: {e}")
```

```python
# Plot V(R) vs R and Cornell Fit
R0_T_MIN = 10
r_min = 2
try:
    r0_var = calc.get_variable("r0", t_min=R0_T_MIN, t_max=t_max_value if 't_max_value' in locals() else None, target_force=SOMMER_TARGET_FORCE, r_min=r_min)
    r0_val = r0_var.get()
    
    cornell_params = r0_var.parameters.get("cornell_params", None)
    
    print(f"Calculated r0: {r0_val}")
    if cornell_params:
        print(f"Cornell Parameters: {cornell_params}")
    else:
        print("Warning: Cornell parameters not found in r0 variable.")

    rs = []
    vs = []
    vs_err = []
    
    for r_val in unique_Rs:
        try:
            v_var = calc.get_variable("V_R", R=r_val, t_min=R0_T_MIN, t_max=t_max_value if 't_max_value' in locals() else None)
            val = v_var.get()
            err = v_var.err()
            
            if not np.isnan(val):
                rs.append(r_val)
                vs.append(val)
                vs_err.append(err if err is not None else 0)
        except Exception:
            continue
            
    fig, ax = plt.subplots(figsize=(8, 6))
    
    ax.errorbar(rs, vs, yerr=vs_err, fmt='o', label='V(R) Data', zorder=5)
    print(f"V(R) data points: {list(zip(rs, vs))}")
    # Plot Cornell Fit Curve
    if cornell_params:
        A = cornell_params['A']
        sigma = cornell_params['sigma']
        B = cornell_params['B']
        
        r_line = np.linspace(min(rs), max(rs), 100)
        v_line = cornell_potential_ansatz(r_line, A, sigma, B)
        
        ax.plot(r_line, v_line, 'r-', label=f'Cornell Fit\nA={A:.2f}, sig={sigma:.2f}, B={B:.2f}')
        
        if not np.isnan(r0_val):
            v_r0 = cornell_potential_ansatz(r0_val, A, sigma, B)
            ax.plot(r0_val, v_r0, 'ks', markersize=8, label=f'$r_0/a$ = {r0_val:.3f}')
            ax.axvline(r0_val, color='k', linestyle=':', alpha=0.5)

    ax.set_title(f"Static Potential V(R) and Cornell Fit (R/a min={r_min})")
    ax.set_xlabel("R/a")
    ax.set_ylabel("aV(R)")
    ax.legend()
    plt.show()

except Exception as e:
    print(f"Error in plotting V(R): {e}")
```

```python
# Plot r0 vs r_min

r_mins_to_test = range(1, 6)
r_mins_success = []
r0_values = []
r0_errors = []

for r_min_val in r_mins_to_test:
    try:
        r0_var_temp = calc.get_variable("r0", t_min=R0_T_MIN, t_max=t_max_value if 't_max_value' in locals() else None, target_force=SOMMER_TARGET_FORCE, r_min=r_min_val)
        r0_values.append(r0_var_temp.get())
        r0_errors.append(r0_var_temp.err() if r0_var_temp.err() is not None else 0)
        r_mins_success.append(r_min_val)
    except Exception as e:
        print(f"Could not calculate r0 for r_min={r_min_val}: {e}")

fig, ax = plt.subplots(figsize=(8, 6))
ax.errorbar(r_mins_success, r0_values, yerr=r0_errors, fmt='s-', capsize=5, color='darkblue', label='$r_0$ vs $r_{min}$')

ax.set_title(f"$r_0$ Stability: Dependence on $r_{{min}}$ (t_min={R0_T_MIN})")
ax.set_xlabel("$r_{min}/a$")
ax.set_ylabel("$r_0/a$")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()

print(f"r0 values: {r0_values}")
print(f"r0 errors: {r0_errors}")
```
